```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
alcohol_file = "russia_alcohol.csv"
alcohol_data_df = pd.read_csv(alcohol_file)
alcohol_data_df.head()
```

```python
start_date = 2000
end_date = 2016

after_start_date = alcohol_data_df["year"] >= start_date
before_end_date = alcohol_data_df["year"] <= end_date
between_dates = after_start_date & before_end_date
alcohol_df = alcohol_data_df.loc[between_dates]

alcohol_df.head()
```

```python
alcohol_df=alcohol_df.set_index('year')

alcohol_df.head()
```

```python
alcohol_df.to_csv('russian_alcohol_df.csv') 
```
